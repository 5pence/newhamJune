```python

import pandas as pd
import matplotlib.pyplot as plt

# Sample data
data = pd.Series([5, 6, 7, 8, 9, 10, 100])  # 100 is the outlier

plt.figure(figsize=(6,1))
plt.boxplot(data, vert=False)
plt.title("Boxplot Showing Outlier")
plt.show()

# Handling outlier by capping (winsorizing)
data_capped = data.copy()
data_capped[data_capped > 20] = 20

plt.figure(figsize=(6,1))
plt.boxplot(data_capped, vert=False)
plt.title("After Handling Outlier (Capped at 20)")
plt.show()
```
