```python
# Training data: 9-pixel images representing "H" and "n"
X = [
    [1, 0, 1,
     1, 1, 1,
     1, 0, 1],  # H

    [0, 1, 0,
     1, 1, 1,
     1, 0, 1]   # n
]

# Labels: H = 1, n = 0
y = [1, 0]

# Initialize weights and bias to 0
weights = [0.0] * 9
bias = 0.0
lr = 0.1  # Learning rate

# Activation function: simple step function
def activate(z):
    return 1 if z > 0 else 0

# Training loop for 5 epochs
for epoch in range(5):
    print(f"\nüîÅ Epoch {epoch + 1}")
    
    # Loop through each training example
    for i, x in enumerate(X):
        print(f"\nExample {i + 1} ‚Äî {'H' if y[i] == 1 else 'n'}")
        print(f"Input: {x}")
        print(f"Current Weights: {weights}")
        print(f"Bias: {bias}")

        # 1. Compute the weighted sum (dot product of weights and inputs + bias)
        z = sum(w * xi for w, xi in zip(weights, x)) + bias
        print(f"Weighted sum (z): {z:.2f}")

        # 2. Apply activation function to get binary prediction
        pred = activate(z)
        print(f"Prediction: {pred} | True: {y[i]}")

        # 3. Calculate error
        error = y[i] - pred
        print(f"Error: {error}")

        # 4. Update weights and bias using the perceptron learning rule
        weights = [w + lr * error * xi for w, xi in zip(weights, x)]
        bias += lr * error

        print(f"Updated Weights: {weights}")
        print(f"Updated Bias: {bias:.2f}")
```
